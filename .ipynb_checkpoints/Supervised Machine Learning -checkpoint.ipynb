```python
#1(d) Let's focus on the following features: danceability, tempo, energy, valence
# for each of these features(in order), produce a histogram that shows the distribution of the feature valuess in training set, separated for positive and negative example. We mean target =1 is positive and by negative we mean target = 0. as an example is what the histogram would look like for a diffrent feature, loudness

features_of_interest = ['danceability', 'tempo', 'energy', 'valence']

positive_target = df_train[df_train['target'] == 1]
negative_target = df_train[df_train['target'] == 0]

# Plot histograms for each feature
for feature in features_of_interest:
    plt.figure(figsize=(8, 6))
    plt.hist(positive_target[feature], bins=30, alpha=0.5, color='b', label='Positive Examples')
    plt.hist(negative_target[feature], bins=30, alpha=0.5, color='r', label='Negative Examples')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.title(f'Distribution of {feature} for Positive and Negative Examples')
    plt.legend()
    plt.grid(True)
    plt.show()
```

# Excercise 2:Using sklearn to bulid a decision tree classifire

```python
#2(a)To create X_train, y_train, X_test, and y_test from df_train and df_test excluding the 
#"song_title" and "artist" features, and then fit a DecisionTreeClassifier

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

features_to_exclude = ["song_title", "artist"]
X_train = df_train.drop(features_to_exclude, axis=1)
y_train = df_train["target"]
X_test = df_test.drop(features_to_exclude, axis=1)
y_test = df_test["target"]

clf = DecisionTreeClassifier()

clf.fit(X_train, y_train)
```

```python
#2(b) Use the predict method to predict the class of first example in your x_train. is the prediction correct? 
#that is does it match with the corresponding class in y_train? hint x_train.iloc[[0]]
prediction = clf.predict(X_train.iloc[[0]])
if prediction == y_train.iloc[0]:
    print("The prediction is correct.")
else:
    print("The prediction does not match the corresponding class in y_train.")
```

```python
#2(C) Use the cross_val_score function on your trainning set to compute the 10-foldcross-validation accuracy of tree.
# from sklearn.model_selection import cross_val_score
# scores = cross_val_score(clf, X_train, y_train, cv=10)

# print(scores)


from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier

Decisiontree = DecisionTreeClassifier()

k=10
kf = KFold(n_splits=k, shuffle=True, random_state=42) #0,7 kuch bhi rakh shakte hai

scores = cross_val_score(Decisiontree, X_train,y_train, cv=kf, scoring='accuracy')

for fold, score in enumerate(scores):
    print(f"Fold {fold+1}: Accuracy = {score:.2f}")
    
mean_accuracy = np.mean(scores)
print(f"Mean Accuracy ={mean_accuracy:.2f}")
```

```python
#2(D)The above is useful,but we would like to see the training accuracy as well 
# compute the 10-fold cross-validation again but this time using the cross_validate function with return_train_score=True. 
# Print out both the cross-validation score and the training score. 
# is your cross validation score exactly the same as what you get in the previous part? 


from sklearn.model_selection import cross_validate
from sklearn.model_selection import KFold
from sklearn.tree import DecisionTreeClassifier

Decisiontree = DecisionTreeClassifier()
kf = KFold(n_splits=10, shuffle=True, random_state=42)

cv_results = cross_validate(Decisiontree, X_train, y_train, cv=kf, return_train_score=True)

print("Cross-validation scores:", cv_results['test_score'])
mean_accuracy =np.mean(cv_results['test_score'])
print(f"Mean Accuracy Cross validation = {mean_accuracy:.2f}")

print("Training scores:", cv_results['train_score'])
mean_accuracy =np.mean(cv_results['train_score'])
print(f"Mean Accuracy Training score = {mean_accuracy:.2f}")
```

```python
#2(e) Do you see a significant difference between the training score and the cross validation score.
#Briefly discuss.
```

Answer 2(e): when the training score and cross-validation score are similar, it suggests that the model is performing consistently well on both the data it was trained on and new, unseen data. 
This is a desirable outcome, indicating that the model has likely found the true underlying patterns in the data.
